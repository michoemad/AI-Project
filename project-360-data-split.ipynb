```python
tweets_19 = pd.read_csv("../input/2019-tweets/2019_detailed_tweets.csv/2019_detailed_tweets.csv", engine='python')
tweets_20 = pd.read_csv("../input/2020-tweets/2020_detailed_tweets.csv/2020_detailed_tweets.csv", engine='python')
```

```python
tweets_19.head()
```

```python
tweets_20.head()
```

```python
tweets_combined = pd.concat([tweets_20, tweets_19])
len(tweets_combined)
```

```python
tweets_combined = tweets_combined.drop_duplicates()
len(tweets_combined)
```

```python
tweets_combined = tweets_combined.loc[tweets_combined['party'] != "party"]
len(tweets_combined)
```

```python
counts = tweets_combined.party.value_counts()
counts
```

```python
selected_dem = tweets_combined.query('party == "D"').sample(counts['R'])
```

```python
balanced_set = pd.concat([tweets_combined.query('party == "R"'), selected_dem], ignore_index=True)
```

```python
train = balanced_set.sample(frac = 0.6)
len(train)
```

```python
non_train = balanced_set.drop(train.index)
len(non_train)
```

```python
test = non_train.sample(frac = 0.5)
len(test)
```

```python
val = non_train.drop(test.index)
len(val)
```

```python
train.to_csv (r'./train_set.csv', index = False, header=True)
val.to_csv (r'./val_set.csv', index = False, header=True)
test.to_csv (r'./test_set.csv', index = False, header=True)
```
