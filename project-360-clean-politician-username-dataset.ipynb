```python
df = pd.read_csv("/kaggle/input/us-politicians-twitter-dataset/dataset.csv")
print(df.shape)
df.head()
```

```python
not_nan_df = df[df['Twitter_username'].notnull()]
print(not_nan_df.shape)
```

```python
cleaned_df = not_nan_df[['Name', 'Twitter_username', 'Political_party']]
cleaned_df.head()
```

```python
dem_df = cleaned_df.loc[cleaned_df['Political_party'] == "Democratic Party"]
print(len(dem_df))
dem_df.head()
```

```python
not_dem_name_set = set(cleaned_df.loc[cleaned_df['Political_party'] != "Democratic Party"]["Name"])
print(len(not_dem_name_set))
```

```python
rep_df = cleaned_df.loc[cleaned_df['Political_party'] == "Republican Party"]
print(len(rep_df))
rep_df.head()
```

```python
not_rep_name_set = set(cleaned_df.loc[cleaned_df['Political_party'] != "Republican Party"]["Name"])
print(len(not_rep_name_set))
```

```python
other_df = cleaned_df.loc[~cleaned_df['Political_party'].isin(["Republican Party", "Democratic Party"])]
print(len(other_df))
other_df.head()
```

```python
# Get only politicians who never switched party
dem_df = dem_df.loc[~dem_df['Name'].isin(not_dem_name_set)]
print(len(dem_df))
dem_df.head()
```

```python
# Get only politicians who never switched party
rep_df = rep_df.loc[~rep_df['Name'].isin(not_rep_name_set)]
print(len(rep_df))
rep_df.head()
```

```python
# Combine 
final_df = pd.concat([dem_df, rep_df])
print(len(final_df))
final_df.head()
```

```python
# Delete duplicate entries containing the same info
final_df = final_df.drop_duplicates()
print(len(final_df))
final_df.head()
```

```python
final_df.to_csv (r'./rep_names.csv', index = False, header=True)
```
